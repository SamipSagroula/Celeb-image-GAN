```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import os

class CelebADataset(Dataset):
    def __init__(self, transform=None):
        """
        Dataset for loading images from the desktop folder.

        Args:
            transform (callable, optional): Optional transform to apply on images.
        """
        # Set the root directory to the desktop folder with images
        self.root_dir = r"C:\Users\samip\OneDrive\Desktop\img_align_celeba"
        self.transform = transform
        
        # Collect all .jpg image file paths from the desktop folder
        self.image_paths = [
            os.path.join(self.root_dir, fname) 
            for fname in os.listdir(self.root_dir) 
            if fname.lower().endswith('.jpg')
        ]

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        
        if self.transform:
            image = self.transform(image)
        
        return image
```
